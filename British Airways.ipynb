```python
# Calculating the eligibility lookup table
"""
Analyzes the flight schedule data to create a lookup table of
average lounge eligibility percentages grouped by Route Haul and Time of Day.
"""
"""
1. Calculate Total Passengers (TOTAL_PAX): We sum the seats in all classes.
    # We use this as the total volume against which we calculate the eligibility percentages.
"""

df = data_content
df["TOTAL_PASS"] = df["FIRST_CLASS_SEATS"] + df["BUSINESS_CLASS_SEATS"] + df["ECONOMY_SEATS"]
```

```python
# 2. Define the Grouping Key (FLIGHT_GROUP):
"""
This combines the two variables we believe influence eligibility most: Haul and Time of Day.
"""
df["FLIGHT_GROUP"] = df["HAUL"] + "-" + df["TIME_OF_DAY"]
df["FLIGHT_GROUP"].head()
```

```python
# 3. The Aggregation and Calculation of Percentages
# Group the entire DataFrame by the new FLIGHT_GROUP key and sum the important columns.
grouped = df.groupby("FLIGHT_GROUP").agg(
    Total_Pax_Volume = ("TOTAL_PASS", "sum"), # sum of all passengers volume(seats)
    Total_Tier1_Eligible = ("TIER1_ELIGIBLE_PAX", "sum"), # sum of all Tier1 eligible Pax in the group
    Total_Tier2_Eligible = ("TIER2_ELIGIBLE_PAX", "sum"), # sum of all Tier2 eligible Pax in the group
    Total_Tier3_Eligible = ("TIER3_ELIGIBLE_PAX", "sum"), # sum of all Tier3 eligible Pax in the group
    Flight_Count = ("FLIGHT_GROUP", "size"), # Count how many flights are in the group.
)
```

```python
 # 4. Calculate the average number of passengers per flight in this group (for context)
grouped["Avg_Pax_Per_Flight"] = (grouped["Total_Pax_Volume"] / grouped["Flight_Count"]).round(0).astype(int)
```

```python
# 5. Calculate the Eligibility Percentage for each tier:
# This is the core assumption: (Total Eligible Pax) / (Total Pax Volume) * 100
grouped["Tier1_Eligibility_Pct"] = (grouped["Total_Tier1_Eligible"] / grouped["Total_Pax_Volume"]) * 100
grouped["Tier2_Eligibility_Pct"] = (grouped["Total_Tier2_Eligible"] / grouped["Total_Pax_Volume"]) * 100
grouped["Tier3_Eligibility_Pct"] = (grouped["Total_Tier3_Eligible"] / grouped["Total_Pax_Volume"]) * 100
```

```python
# Select the columns that form the final, clean lookup table
lookup_table = grouped[[
        "Avg_Pax_Per_Flight",
        "Tier1_Eligibility_Pct",
        "Tier2_Eligibility_Pct",
        "Tier3_Eligibility_Pct"
    ]].sort_index()  # Sort by the grouping name (A-Z)
```

```python
def create_lookup_table(data_content):
    """
    Function to create a lookup table for eligibility percentages.
    The total volume against which we calculate the eligibility percentages.
    """

    df = data_content
    df["TOTAL_PASS"] = df["FIRST_CLASS_SEATS"] + df["BUSINESS_CLASS_SEATS"] + df["ECONOMY_SEATS"]

    # 2. Define the Grouping Key (FLIGHT_GROUP):
    """
    This combines the two variables we believe influence eligibility most: Haul and Time of Day.
    """
    df["FLIGHT_GROUP"] = df["HAUL"] + "-" + df["TIME_OF_DAY"]

    # 3. The Aggregation and Calculation of Percentages
    # Group the entire DataFrame by the new FLIGHT_GROUP key and sum the important columns.
    grouped = df.groupby("FLIGHT_GROUP").agg(
        Total_Pax_Volume = ("TOTAL_PASS", "sum"), # sum of all passengers volume(seats)
        Total_Tier1_Eligible = ("TIER1_ELIGIBLE_PAX", "sum"), # sum of all Tier1 eligible Pax in the group
        Total_Tier2_Eligible = ("TIER2_ELIGIBLE_PAX", "sum"), # sum of all Tier2 eligible Pax in the group
        Total_Tier3_Eligible = ("TIER3_ELIGIBLE_PAX", "sum"), # sum of all Tier3 eligible Pax in the group
        Flight_Count = ("FLIGHT_GROUP", "size"), # Count how many flights are in the group.
    )

    # 4. Calculate the average number of passengers per flight in this group (for context)
    grouped["Avg_Pax_Per_Flight"] = (grouped["Total_Pax_Volume"] / grouped["Flight_Count"]).round(0).astype(int)

    # 5. Calculate the Eligibility Percentage for each tier:
    # This is the core assumption: (Total Eligible Pax) / (Total Pax Volume) * 100
    grouped["Tier1_Eligibility_Pct"] = (grouped["Total_Tier1_Eligible"] / grouped["Total_Pax_Volume"]) * 100
    grouped["Tier2_Eligibility_Pct"] = (grouped["Total_Tier2_Eligible"] / grouped["Total_Pax_Volume"]) * 100
    grouped["Tier3_Eligibility_Pct"] = (grouped["Total_Tier3_Eligible"] / grouped["Total_Pax_Volume"]) * 100

    # Select the columns that form the final, clean lookup table
    lookup_table = grouped[[
            "Avg_Pax_Per_Flight",
            "Tier1_Eligibility_Pct",
            "Tier2_Eligibility_Pct",
            "Tier3_Eligibility_Pct"
        ]].sort_index()  # Sort by the grouping name (A-Z)

    return lookup_table
```
